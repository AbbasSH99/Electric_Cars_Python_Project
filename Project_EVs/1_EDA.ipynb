```python
import pandas as pd

df = pd.read_excel('../Original_data/Electric_Vehicle_Data.xlsx')
```

```python
overview = df.info(buf=None)  # buf=None prints directly, change to a variable to capture as text

# If you want to capture the overview as a string to print or use elsewhere:
import io

buffer = io.StringIO()
df.info(buf=buffer)
overview_text = buffer.getvalue()

# Print the overview as text
print(overview_text)
```
