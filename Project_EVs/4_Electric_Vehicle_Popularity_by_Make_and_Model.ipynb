# Which makes and models of electric vehicles are most popular?

### Methedology

1. Group by Make and count the occurrences
2. Group by both Make and Model to count the number of each model
3. Aggregate the data and find out which makes and models are the most common
4. Visualize the results to easily identify the most popular makes and models

```python
# Importing libraries and loading the data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('../Original_data/Electric_Vehicle_Data.xlsx')
```

### Grouping the Data by Make and Model

```python
# Count the number of vehicles for each make
popular_makes = df.groupby('Make').size().reset_index(name='Count').sort_values(by='Count', ascending=False).reset_index(drop=True).copy()

# Count the number of vehicles for each make and model combination
popular_models = df.groupby(['Make', 'Model']).size().reset_index(name='Count').sort_values(by='Count', ascending=False).reset_index(drop=True).copy()
```

```python
popular_models
```

### Visualizing the results

```python
# Plot the most popular makes
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Make', data=popular_makes.head(10), palette='Blues_d')
plt.title('Top 10 Most Popular Electric Vehicle Makes')
plt.xlabel('Number of Vehicles')
plt.ylabel('Make')
plt.show()

# Ensure both 'Make' and 'Model' are treated as strings before concatenation
popular_models['Make'] = popular_models['Make'].astype(str)
popular_models['Model'] = popular_models['Model'].astype(str)

# Create a new column that combines Make and Model
popular_models['Make_Model'] = popular_models['Make'] + ' (' + popular_models['Model'] + ')'

# Plot the most popular models with combined Make and Model labels
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Make_Model', data=popular_models.head(10), palette='Blues_d')
plt.title('Top 10 Most Popular Electric Vehicle Models')
plt.xlabel('Number of Vehicles')
plt.ylabel('Make (Model)')
plt.show()
```
