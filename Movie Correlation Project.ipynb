```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data

df = pd.read_csv(r"C:\Users\ACER\Desktop\Data Analysis\Data Analysis with Python\Projects Data Set\movies.csv")
```

```python
# Let's have a look at our data

df.head()
```

```python
#Checking for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Checking the data types for the columns

df.dtypes
```

```python
# Modifying data types to get rid of unecessary decimal place amd filling empyt columns with number 0.

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
# Correcting the year column.

df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
```

```python
df.head()
```

```python
df = df.sort_values(by='gross', inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Dropping duplicates to check for possible flaws in the data

df['company'].drop_duplicates().sort_values(ascending=False).head(20)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show
```

```python
# Converting object types to numeric so we can visualise correlation more effectively

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes
        
df_numerized.head(10)
```

```python
# A quik look to compare the data frames

df.head(10)
```

```python
# Looking at the updated heatmap

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Unstacking for a better view of correlation

correlation_mat = df_numerized.corr()

correlation_pairs = correlation_mat.unstack()

correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()

sorted_pairs.head(15)
```

```python
# Looking at pairs with high correlation

high_corr = sorted_pairs[(sorted_pairs)  >= 0.5]

high_corr
```
